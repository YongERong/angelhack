```python
import requests
import pdfplumber
import io 


def extract_pdf(link):
    url = link
    response = requests.get(url)
    text = ""
    with pdfplumber.open(io.BytesIO(response.content)) as pdf:
        for page in pdf.pages:
            text += page.extract_text() + "\n"
    return text

pdf_text = extract_pdf("https://www.moneysense.gov.sg/files/Basic%20Financial%20Planning%20Guide/english__fresh_entrant.pdf")
print(pdf_text)
```
